```python
# -------------------------------------------
# Step 1: Install required packages (Colab)
# -------------------------------------------
!pip install pandas matplotlib seaborn

# -------------------------------------------
# Step 2: Import libraries
# -------------------------------------------
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots look nice
sns.set(style="whitegrid")

# -------------------------------------------
# Step 3: Load dataset
# (Using a free public dataset so no upload needed)
# -------------------------------------------
url = "https://raw.githubusercontent.com/datasets/population/master/data/population.csv"
df = pd.read_csv(url)

# Show first few rows
df.head()
```

```python
# -------------------------------------------
# Step 4: Basic data check
# -------------------------------------------
print("Dataset shape:", df.shape)
print("\nColumn names:", df.columns)
print("\nMissing values per column:\n", df.isnull().sum())

# Filter for most recent year available
latest_year = df["Year"].max()
df_latest = df[df["Year"] == latest_year]

df_latest.head()
```

```python
# -------------------------------------------
# Step 5: Top 10 countries by population
# -------------------------------------------
top10 = df_latest.sort_values(by="Value", ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(data=top10, x="Value", y="Country Name", palette="viridis")
plt.xlabel("Population")
plt.ylabel("Country")
plt.title(f"Top 10 Countries by Population ({latest_year})")
plt.tight_layout()
plt.show()
```

```python
# -------------------------------------------
# Step 6: Population growth of a specific country
# -------------------------------------------
country = "India"  # You can change this
country_data = df[df["Country Name"] == country]

plt.figure(figsize=(10, 6))
sns.lineplot(data=country_data, x="Year", y="Value", marker="o")
plt.xlabel("Year")
plt.ylabel("Population")
plt.title(f"Population Growth of {country}")
plt.tight_layout()
plt.show()
```

```python
# -------------------------------------------
# Step 7: Save cleaned data for GitHub
# -------------------------------------------
df_latest.to_csv("world_population_latest.csv", index=False)
print("Cleaned dataset saved as 'world_population_latest.csv'")
```
